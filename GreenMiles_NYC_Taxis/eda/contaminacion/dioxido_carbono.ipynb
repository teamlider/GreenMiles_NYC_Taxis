```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Creación del dataframe df_co2

```python
df_co2 = pd.read_csv('energy.csv')
df_co2.head(10)
```

## Diccionario de datos. 'energy.csv
### Country: El nombre del país al que pertenecen los datos.
### Energy_type: El tipo de energía asociado con los datos.
### Year: El año al que corresponden los datos.
### Energy_consumption: El consumo de energía, medido en (quad Btu) "quadrillion British Thermal Units" (un cuadrillón de unidades térmicas británicas).
### Energy_production: La producción de energía, medido (quad Btu)
### GDP: Producto Interno Bruto (Gross Domestic Product), medida en cifras en miles de millones de dólares estadounidenses ajustados por paridad de poder adquisitivo al año 2015. ( 2015 año de referencia para estos ajustes).
### Population: Número de habitantes en el país. medida (Mperson).
### Energy_intensity_per_capita: La intensidad energética per cápita, que puede ser una medida de cuánta energía se consume por persona,Se calcula como unidades de energía por unidad de habitante (capita = persona individual), medida (MMBtu/persona).
### Energy_intensity_by_GDP: La intensidad energética en relación con el Producto Interno Bruto (GDP).Se calcula como unidades de energía por unidad de PIB, medido (1000 Btu/2015$ PIB PPA).
### CO2_emission: Las emisiones de dióxido de carbono asociadas con la actividad energética y económica. medida (MMtoneladas de CO2).

###  Filtro las filas con los valores de Energy_type 'natural_gas', 'petroleum_n_other_liquids,all_energy_types'
###  Selecciono solo las columnas 'Country', 'Energy_type', 'Year', 'Energy_consumption', 'Population', 'CO2_emission'
### Creo una nueva tabla con los datos filtrados.
### El objetivo de filtrar estos datos es poder analizar las emisiones de C02 producto del consumo de combustible fosiles y gas natural en comparación con todos los tipos de energia y que porcentanjes representan. tanto en el consumo como en la generación de CO2.

```python

filtered_df = df_usa[df_usa['Energy_type'].isin(['natural_gas', 'petroleum_n_other_liquids','all_energy_types'])]

selected_columns = ['Country', 'Energy_type', 'Year', 'Energy_consumption', 'Population', 'CO2_emission']
result_table = filtered_df[selected_columns]

result_table.info()
```

```python
result_table.head()
```

### Creo una tabla para poder observar la generación de CO2 con todos los tipos de energia desde el año 1980 al 2020.

```python

all_energy_types_df = result_table[result_table['Energy_type'] == 'all_energy_types']

co2_by_year_all_types = all_energy_types_df.groupby('Year')['CO2_emission'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(co2_by_year_all_types['Year'], co2_by_year_all_types['CO2_emission'], color='skyblue')
plt.xlabel('Año')
plt.ylabel('Generación Total de CO2 (Millones de Toneladas)')
plt.title('Generación Total de CO2 para all_energy_types por Año')
plt.show()
```

###  Creo una gráfica de línea de tendencia. 

```python


all_energy_types_df = result_table[result_table['Energy_type'] == 'all_energy_types']


co2_by_year_all_types = all_energy_types_df.groupby('Year')['CO2_emission'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.regplot(x='Year', y='CO2_emission', data=co2_by_year_all_types, ci=None, line_kws={'color': 'red'})
plt.xlabel('Año')
plt.ylabel('Generación Total de CO2 (Millones de Toneladas)')
plt.title('Tendencia de Generación Total de CO2 para all_energy_types por Año')
plt.show()
```

### Creo una gráfica de dispersión para ver la producción de CO2 a través de los años producto del natural_gas.

```python

natural_gas_df = result_table[result_table['Energy_type'] == 'natural_gas']

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='CO2_emission', data=natural_gas_df, color='green')
plt.xlabel('Año')
plt.ylabel('Generación de CO2 (Millones de Toneladas)')
plt.title('Generación de CO2 por natural_gas por Año')
plt.show()
```

```python

selected_energy_types_df = result_table[result_table['Energy_type'].isin(['natural_gas', 'petroleum_n_other_liquids'])]

plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.scatterplot(x='Year', y='CO2_emission', data=selected_energy_types_df[selected_energy_types_df['Energy_type'] == 'natural_gas'], color='green')
plt.title('Generación de CO2 por natural_gas')

plt.subplot(1, 2, 2)
sns.scatterplot(x='Year', y='CO2_emission', data=selected_energy_types_df[selected_energy_types_df['Energy_type'] == 'petroleum_n_other_liquids'], color='blue')
plt.title('Generación de CO2 por petroleum_n_other_liquids')

plt.tight_layout()
plt.show()
```

### Creo varias tortas en donde se expone los porcentajes de producción de CO2 por 'natural_gas' y petroleum_n_other_liquids. en los diferentes años.

```python

selected_energy_types_df = result_table[result_table['Energy_type'].isin(['natural_gas', 'petroleum_n_other_liquids'])]


years = selected_energy_types_df['Year'].unique()
num_years = len(years)

plt.figure(figsize=(15, 5 * num_years))

for i, year in enumerate(years, 1):
    plt.subplot(num_years, 2, i)
    
 
    year_df = selected_energy_types_df[selected_energy_types_df['Year'] == year]
    
  
    energy_percentage = year_df.groupby('Energy_type')['CO2_emission'].sum() / year_df['CO2_emission'].sum() * 100
    

    plt.pie(energy_percentage, labels=energy_percentage.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
    plt.title(f'Producción de CO2 por tipo de energía - Año {year}')

plt.tight_layout()
plt.show()
```

### Creo una tabla pivot con la misma información de las tortas

```python

selected_energy_types_df = result_table[result_table['Energy_type'].isin(['natural_gas', 'petroleum_n_other_liquids'])]

pivot_df = selected_energy_types_df.pivot(index='Year', columns='Energy_type', values='CO2_emission').fillna(0)

pivot_df_percentage = pivot_df.divide(pivot_df.sum(axis=1), axis=0) * 100


table_df = pd.DataFrame(pivot_df_percentage)
table_df.columns.name = 'Tipo de Energía'
table_df.index.name = 'Año'

table_df.head(21)
```
