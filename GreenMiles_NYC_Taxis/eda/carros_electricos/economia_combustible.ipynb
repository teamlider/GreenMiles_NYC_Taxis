**Year**: Año de fabricación del vehículo.  
**Manufacturer** Fabricante del vehículo.  
**Model:** Modelo del vehículo.  
**barrels08:** Consumo de barriles de combustible en ciudad por cada 100 millas.  
**barrelsA08:** Consumo de barriles de combustible en carretera por cada 100 millas.  
**charge240:** Tiempo de carga de la batería eléctrica en un cargador de 240V.    
**city08:** Consumo de combustible en ciudad en millas por galón.  
**city08U:** Consumo de combustible en ciudad en millas por galón (ajustado).  
**cityA08:** Consumo de combustible en ciudad y en modo eléctrico en millas por galón.  
**cityA08U:** Consumo de combustible en ciudad y en modo eléctrico ajustado en millas por galón.  
**co2:** Emisiones de dióxido de carbono en gramos por milla.  
**co2A:** Emisiones de dióxido de carbono ajustadas en gramos por milla.  
**co2TailpipeAGpm:** Emisiones de dióxido de carbono de la cola del tubo de escape en gramos por milla.  
**co2TailpipeGpm**: Emisiones de dióxido de carbono de la tubería de escape en gramos por milla.  
**comb08:** Consumo de combustible combinado en millas por galón.  
**comb08U:** Consumo de combustible combinado en millas por galón (ajustado).  
**combA08:** Consumo de combustible combinado en modo eléctrico en millas por galón.  
**combA08U:** Consumo de combustible combinado en modo eléctrico ajustado en millas por galón.  
**cylinders:** Número de cilindros del motor.  
**displ:** Desplazamiento del motor en litros.  
**drive:** Tipo de tracción del vehículo (Rear-Wheel Drive, Front-Wheel Drive, etc.).  
**engId:** Identificador único del motor.  
**eng_dscr:** Descripción del motor.  
**feScore:** Puntuación de eficiencia de combustible del vehículo.   
**fuelCost08:** Costo estimado de combustible en ciudad para 100 millas.   
**fuelCostA08:** Costo estimado de combustible en carretera para 100 millas.  
**fuelType:** Tipo de combustible principal del vehículo.  
**fuelType1:** Tipo de combustible secundario del vehículo.  
**ghgScore:** Puntuación de emisiones de gases de efecto invernadero.  
**ghgScoreA:** Puntuación ajustada de emisiones de gases de efecto invernadero.  
**highway08:** Consumo de combustible en carretera en millas por galón.  
**highway08U:** Consumo de combustible en carretera en millas por galón (ajustado).  
**highwayA08:** Consumo de combustible en carretera y en modo eléctrico en millas por galón.  
**highwayA08U:** Consumo de combustible en carretera y en modo eléctrico ajustado en millas por galón.  
**VClass:** Clase del vehículo (Subcompact Cars, Midsize Cars, etc.).  
**highwayCD:** Consumo de combustible en carretera en galones por 100 millas.  
**highwayE:** Consumo de combustible en carretera en modo eléctrico en millas por galón.  
**highwayUF:** Consumo de combustible en carretera en modo eléctrico ajustado en millas por galón.  
**hlv:** Volumen útil de carga pesada del vehículo.  
**hpv:** Peso de carga pesada del vehículo.  
**id:** Identificador único del vehículo.  
**lv2:** Nivel de emisiones en ciudad.  
**lv4:** Nivel de emisiones en carretera.  
**mpgData:** Indica si se dispone de datos de consumo de combustible en millas por galón.  
**phevBlended:** Indica si el vehículo es una versión híbrida enchufable.  
**pv2:** Puntuación de vehículos en ciudad.  
**pv4:** Puntuación de vehículos en carretera.  
**range:** Rango total del vehículo.  
**rangeCity:** Rango del vehículo en ciudad.  
**rangeCityA:** Rango ajustado del vehículo en ciudad.  
**rangeHwy:** Rango del vehículo en carretera.  
**rangeHwyA:** Rango ajustado del vehículo en carretera.  
**trany:** Tipo de transmisión del vehículo.  
**UCity:** Consumo de combustible en ciudad ajustado en galones por 100 millas.  
**UCityA:** Consumo de combustible en ciudad y en modo eléctrico ajustado en galones por 100 millas.  
**UHighway:** Consumo de combustible en carretera ajustado en galones por 100 millas.  
**UHighwayA:** Consumo de combustible en carretera y en modo eléctrico ajustado en galones por 100 millas.  
**youSaveSpend:** Ahorro o gasto estimado en combustible durante 5 años en comparación con el promedio del vehículo.  
**guzzler:** Indica si el vehículo es un "guzzler" (consumidor voraz de combustible).  
**trans_dscr:** Descripción de la transmisión del vehículo.  
**tCharger:** Indica si el vehículo tiene un cargador térmico.  
**sCharger:** Indica si el vehículo tiene un cargador solar.  
**atvType:** Tipo de vehículo alternativo.  
**fuelType2:** Tipo de combustible secundario del vehículo.  
**rangeA:** Rango ajustado total del vehículo.  
**evMotor:** Tipo de motor eléctrico.  
**mfrCode:** Código del fabricante del vehículo.  
**c240Dscr:** Descripción del motor C240.  
**charge240b:** Tiempo de carga de la batería eléctrica en un cargador de 240V (segunda batería).  
**c240bDscr:** Descripción de la segunda batería del motor C240.  
**createdOn:** Fecha de creación de los datos.  
**modifiedOn:** Fecha de modificación de los datos.  
**startStop:** Indica si el vehículo tiene la función de arranque/parada automática.  
**phevCity:** Consumo de combustible en modo eléctrico en ciudad.  
**phevHwy:** Consumo de combustible en modo eléctrico en carretera.  
**phevComb:** Consumo de combustible en modo eléctrico combinado.
