### Diccionario

1. **Brand:** Marca del vehículo.
2. **Model:** Modelo específico del vehículo.
3. **AccelSec:** Aceleración de 0 a 100 km/h en segundos.
4. **TopSpeed_KmH:** Velocidad máxima en kilómetros por hora.
5. **Range_Km:** Rango de conducción en kilómetros.
6. **Efficiency_WhKm:** Eficiencia energética en vatios-hora por kilómetro.
7. **FastCharge_KmH:** Velocidad de carga rápida en kilómetros por hora.
8. **RapidCharge:** Capacidad de carga rápida (Sí/No).
9. **PowerTrain:** Tipo de tren motriz (tracción delantera, tracción trasera, tracción a las cuatro ruedas).
10. **PlugType:** Tipo de conector de carga.
11. **BodyStyle:** Estilo de carrocería del vehículo.
12. **Segment:** Segmento al que pertenece el vehículo.
13. **Seats:** Número de asientos en el vehículo.
14. **PriceEuro:** Precio del vehículo en euros.

### Segun Brand: Marca del vehículo

```python
# Graficar la distribución de marcas
plt.figure(figsize=(12, 6))
electric_car_data['Brand'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribución de Marcas de Vehículos Eléctricos')
plt.xlabel('Marca')
plt.ylabel('Número de Vehículos')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.show()
```

Al examinar la información, se destaca que Tesla lidera en términos de número de vehículos eléctricos, con un total de 13 unidades. A continuación, marcas como Audi, Nissan y Volkswagen siguen de cerca con 9, 8 y 8 unidades respectivamente, indicando su presencia significativa en el conjunto de datos.

La variedad de marcas es evidente, ya que se observa la inclusión de marcas conocidas y establecidas, como Tesla y Audi, así como marcas menos comunes que tienen solo una unidad. Esta diversidad sugiere una representación amplia de fabricantes de vehículos eléctricos en el conjunto de datos.

En términos de popularidad, la cantidad de vehículos eléctricos asociados a cada marca puede considerarse un indicador. Tesla destaca como la marca más popular en este conjunto de datos, una posición esperada dada su posición destacada en el mercado de vehículos eléctricos. La concentración de vehículos eléctricos en pocas marcas importantes, como Tesla, Audi y Nissan, es notable, mientras que otras marcas tienen una presencia más limitada.

```python
# Configuración del tamaño del gráfico
plt.figure(figsize=(8, 6))

# Crear un boxplot de la columna 'AccelSec'
sns.boxplot(x=electric_car_data['AccelSec'], color='skyblue')

# Configuración de etiquetas y título
plt.title('Diagrama de Caja para la Aceleración (AccelSec)')
plt.xlabel('Aceleración (segundos)')

# Mostrar el gráfico
plt.show()
```

El conjunto de datos de la columna AccelSec consta de 103 observaciones de tiempo de aceleración en segundos. La estadística descriptiva proporciona una visión clara de la distribución de estos tiempos. La aceleración promedio es de aproximadamente 7.4 segundos, con una dispersión alrededor de esta media de aproximadamente 3.02 segundos, indicada por la desviación estándar.

En términos de cuartiles, el 25% de las observaciones tienen una aceleración de 5.1 segundos o menos, la mediana (50%) es de 7.3 segundos, y el 75% de las observaciones tienen una aceleración de 9.0 segundos o menos. Este rango intercuartílico (IQR) de 3.9 segundos refleja la variabilidad intercuartílica en la distribución.

El valor mínimo de aceleración es de 2.1 segundos, mientras que el valor máximo es notablemente más alto, alcanzando los 22.4 segundos. Este valor máximo, que está fuera de los "bigotes" del diagrama de caja, sugiere la presencia de un valor atípico significativamente más alto que el resto de las observaciones y puede requerir una investigación más detallada para comprender la razón detrás de este extremo inusual, la referecia seria que pasa de 0 a 100 kms/h en 22.4 segundos.

```python
# Configuración del tamaño del gráfico
plt.figure(figsize=(8, 6))

# Crear un gráfico de barras para 'PowerTrain'
electric_car_data['PowerTrain'].value_counts().plot(kind='bar', color='skyblue')

# Configuración de etiquetas y título
plt.xlabel('Tipo de Tren Motriz')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Tipos de Tren Motriz')

# Mostrar el gráfico
plt.show()
```

La distribución de tipos de tren motriz en el conjunto de datos revela que la tracción a las cuatro ruedas (AWD) es la configuración más prevalente, con 41 observaciones. Esta preferencia sugiere una inclinación hacia vehículos que ofrecen una tracción más equilibrada en diversas condiciones de manejo. La tracción delantera (FWD) sigue de cerca con 37 observaciones, indicando una considerable presencia de vehículos con eficiencia de combustible y estabilidad. La tracción trasera (RWD) es menos común pero aún está presente en 25 observaciones, sugiriendo una presencia de vehículos de alto rendimiento en el conjunto de datos.

En resumen, la distribución refleja una variedad de preferencias en términos de tipos de tren motriz, destacando la popularidad de la tracción a las cuatro ruedas en los vehículos eléctricos estudiados.

```python
# Configuración del tamaño del gráfico
plt.figure(figsize=(10, 6))

# Crear un gráfico de barras para 'PlugType'
sns.countplot(x='PlugType', data=electric_car_data, order=electric_car_data['PlugType'].value_counts().index, palette='viridis')

# Configuración de etiquetas y título
plt.xlabel('Tipo de Conector de Carga')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Tipos de Conector de Carga')

# Rotar las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.show()
```

El análisis de la distribución de tipos de conectores de carga en el conjunto de datos revela patrones significativos. En primer lugar, destaca que el conector Type 2 CCS es abrumadoramente predominante, presentándose en un total de 90 observaciones. Esta abrumadora mayoría sugiere que la gran mayoría de los vehículos eléctricos en el conjunto de datos están equipados con el conector Type 2 CCS, lo que puede reflejar una preferencia generalizada en la infraestructura de carga o en las elecciones de diseño de los fabricantes de vehículos.

Por otro lado, el conector Type 2 ocupa el segundo lugar en términos de frecuencia, aunque con una presencia mucho menor que el Type 2 CCS, contabilizando solo 9 observaciones. Esto podría indicar que, aunque menos común, aún hay un segmento de vehículos eléctricos que utilizan este tipo de conector.

En contraste, los conectores Type 2 CHAdeMO y Type 1 CHAdeMO son menos comunes en el conjunto de datos, con 3 y 1 observaciones respectivamente. Estos resultados sugieren una presencia limitada de estos tipos de conectores en comparación con los dos tipos principales.
